```python
#date, msg id, first name, phone, user id, content

total_count, messages, senders = client.get_message_history(chat, limit=10)
last = messages[-1].id

i = 1
last_ = 0
tries = 0
msgs = {}
while 1:
    try:
        total_count, messages, senders = client.get_message_history(chat, offset_id= 250*i, limit=250)
    except:
        sleep(10)
        total_count, messages, senders = client.get_message_history(chat, offset_id= 250*i, limit=250)
    
    for n in range(len(messages)):
        msg = messages[n]
        sender = senders[n]
        send_ = {'name':sender.first_name, 
                 'username':sender.username,
                'phone':sender.phone}
        
        msg.__dict__.update(send_)
        msgs.update({msg.id:{k:str(v) for k,v in msg.__dict__.items()}})
        
    if len(messages) < 1:
        continue
    print('Fetched', len(msgs), 'messages.')
    last = messages[-1].id
    if last_ == last:
        tries += 1
    if tries > 5:
        break
    last_ = last
    i += 1
```

```python
with open('MensajesLPjson.txt', 'w') as out:
    #reverse 
    out.write(json.dumps(msgs))
```
